# Generando archivos PDF del presupuesto

* Generar el archivo PDF
* Definir un tipo de letra
* llenamos el documento
* Guardar el entregable

```python
%pip install fpdf2
```

```python
from fpdf import FPDF
```

```python
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.cell(200, 10, txt="Presupuesto", ln=True, align="C")
pdf.cell(200, 10, txt="Presupuesto", ln=True, align="C")
pdf.cell(200, 10, txt="Presupuesto", ln=True, align="C")
pdf.cell(200, 10, txt="Presupuesto", ln=True, align="C")
pdf.output("Presupuesto.pdf")
```

```python
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.image("Template.png", 0, 0)
pdf.text(115, 145, "Presupuesto")
pdf.text(115, 160, str(Horas_estimadas))
pdf.text(115, 175, str(valor_hora))
pdf.text(115, 190, str(plazo_estimado))
pdf.text(115, 205, str(Costo_total))
pdf.output("Presupuesto.pdf")
```
